```python
from keras.preprocessing import image
testImage='/content/drive/MyDrive/PFE/Detection Model/Test Images/i4.jpeg'
test_image=image.load_img(testImage,target_size=(128, 128))
test_image=image.img_to_array(test_image)
test_image=np.expand_dims(test_image,axis=0)
result=model.predict(test_image,verbose=0)
```

```python
def fruitName(result):

  if np.argmax(result) == 0:
      fruit = 'Apple'
  elif np.argmax(result)==1:
      fruit = 'Banana'
  elif np.argmax(result)==2:
      fruit = 'Cherry'
  elif np.argmax(result)==3:
      fruit = 'Grape'
  elif np.argmax(result)==4:
      fruit = 'Kiwi'
  elif np.argmax(result)==5:
      fruit = 'Mango'
  elif np.argmax(result)==6:
      fruit = 'Orange'
  elif np.argmax(result)==7:
      fruit = 'Peach'
  elif np.argmax(result)==8:
      fruit = 'Pear'
  elif np.argmax(result)==9:
      fruit = 'Plum'
  elif np.argmax(result)==10:
      fruit = 'Strawberry'
  else:
      fruit = 'Watermelon'
  return fruit
```
